```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
data=pd.read_excel('Telco-Customer-Churn.xlsx')
```

```python
data.head()
```

```python
data.tail()
```

```python
data['MultipleLines'].value_counts()
```

```python
data['OnlineBackup']=data['OnlineBackup'].replace({'No internet service': 'No'})
```

```python
data['OnlineSecurity']=data['OnlineSecurity'].replace({'No internet service': 'No'})
```

```python
data['MultipleLines']=data['MultipleLines'].replace({'No phone service': 'No'})
```

```python
data['MultipleLines'].value_counts()
```

```python
data['Churn'].value_counts()
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python


for col in data.select_dtypes(include=['object']).columns:
    data[col] = data[col].astype(str)  # ensure uniform data type
    data[col] = encoder.fit_transform(data[col])
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
x=data.drop(['Churn'], axis=1)
y=data['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=123)
```

```python
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(x_train,y_train)
```

```python
y_knn_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay,precision_score,recall_score,f1_score
```

```python
accuracy_score(y_test, y_knn_pred)
```

```python
precision_score(y_test, y_knn_pred)
```

```python
recall_score(y_test, y_knn_pred)
```

```python
f1_score(y_test, y_knn_pred)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_knn_pred)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
navie = GaussianNB()
```

```python
navie.fit(x_train, y_train)
```

```python
y_navie_pred = navie.predict(x_test)
```

```python
accuracy_score(y_test, y_navie_pred)
```

```python
precision_score(y_test, y_navie_pred)
```

```python
f1_score(y_test, y_navie_pred)
```

```python
recall_score(y_test, y_navie_pred)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_navie_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision_model=DecisionTreeClassifier(max_depth=4)
```

```python
decision_model.fit(x_train,y_train)
```

```python
y_pred=decision_model.predict(x_test)
```
